```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelEncoder
```

```python
fraud = pd.read_csv('Fraud_check.csv')

fraud
```

```python
fraud['Taxable.Income'] = np.where((fraud['Taxable.Income']<=30000), 'Risky', 'Good')

fraud
```

```python
lb = LabelEncoder()
```

```python
fraud['Taxable.Income']= lb.fit_transform(fraud['Taxable.Income'])


fraud = pd.get_dummies(fraud, columns=['Undergrad','Marital.Status','Urban'], drop_first = True)
```

```python
fraud
```

```python
def norm_func(i):
    x= (i-i.min())/(i.max()-i.min())
    return(x)
```

```python
fraud2 = norm_func(fraud)


fraud2
```

```python
x = fraud2.iloc[:,1:]
y = fraud2.iloc[:,0]
```

```python
x
```

```python
y
```

```python
# BUILDING RANDOM FOREST CLASSIFIER MODEL
```

```python
kfold = KFold(n_splits=10, random_state=2, shuffle=True)

model = RandomForestClassifier(n_estimators=200,max_features=3)

results = cross_val_score(model,x,y,cv=kfold)


results.mean()
```
