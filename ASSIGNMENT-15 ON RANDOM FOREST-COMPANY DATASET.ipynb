```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
company = pd.read_csv('Company_Data.csv')
```

```python
company
```

```python
company['highsales'] = np.where((company['Sales']<9), 'low','high')
company
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()


company['highsales'] = lb.fit_transform(company['highsales'])
company['ShelveLoc'] = lb.fit_transform(company['ShelveLoc'])
company['Urban'] = lb.fit_transform(company['Urban'])
company['US'] = lb.fit_transform(company['US'])



company
```

```python
company2 = company.iloc[:,1:]
company2
```

```python
x = company2.iloc[:,:10]
y= company2.iloc[:,10]
```

```python
x
```

```python
y
```

```python
#BUILDING RANDOM FOREST CLASSIFIER MODEL
```

```python
kfold = KFold(n_splits=10,random_state=1,shuffle=True)
model = RandomForestClassifier(n_estimators=40, max_features=4)

```

```python
result = cross_val_score(model,x,y,cv = kfold)
```
